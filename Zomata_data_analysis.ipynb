# **Zomato Data Analysis Project**

# **Import Libraries**

```python
'''panads is used for data manipulation and analysis.
numpy is used for mathematical operations.
matplotlib and seaborn are used for data visualization.'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Create the data frame**

```python
dataframe = pd.read_csv('Zomato data .csv')
print(dataframe)
```

```python
dataframe
```

# **Covert the data type of column - rate**

```python
def handleRate(value):
  value = str(value).split('/')
  value = value[0]
  return float(value)

dataframe['rate']=dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

# **Check for null values**

```python
dataframe.info()
```

# **Type of restaurant from which majority of customers order from**

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'], color="cyan")
plt.xlabel("type of restaurant")
```

# **Conclusion - majority of the restaurant falls in dinning category**

```python
dataframe.head()
```

# **Restaurant type and their voting**

```python
grouped_data = dataframe.groupby('listed_in(type)')['votes'].sum()
result = pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green", marker = 'o', ms = 20, mec = 'g', mfc = 'hotpink')
plt.xlabel("Type of restaurant", c="g", size=20)
plt.ylabel("Votes", c="g", size=20)
```

# **Conclusion - dinning restaurants have recieved maximum votes**

```python
dataframe.head()
```

# **Majority rating**

```python
plt.hist(dataframe['rate'],bins=6,color="y")
plt.title("Ratings Distribution", c="r", size=20)
plt.show()
```

# **Conclusion - majority restaurants recieved ratings from 3.5 to 4**

```python
dataframe.head()
```

# **Average spending by couples on food orders**

```python
couple_data=dataframe['approx_cost(for two people)']
sns.countplot(x=couple_data, color="magenta")
```

# **Conclusion - majority couples prefer restaurants with an approximate cost of Rs. 300**

```python
dataframe.head()
```

# **Which mode recieves maximum rating - online/offline**

```python
plt.figure(figsize=(6,6))
sns.boxplot(x = 'online_order', y = 'rate', data = dataframe)
```

# **Conclusion - online orders recieved higher ratings than offline orders**

```python
dataframe.head()
```

# **Which type of restaurants need good offers by Zomato to promote online orders**

```python
pivot_table = dataframe.pivot_table(index = 'listed_in(type)', columns = 'online_order', aggfunc = 'size', fill_value=0)
sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='d')
plt.title("Heatmap", c='red', size=20)
plt.xlabel("Online Order")
plt.ylabel("Listed In(Type)")
plt.show()
```

# **Conclusion - dinning restaurants primarily accepts offline orders, whereas cafes primarily recieve online orders. This suggests that clients prefer orders in person at restaurants, but prefer online ordering at cafes**
